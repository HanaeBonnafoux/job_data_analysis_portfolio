```python
#setup

import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np

print("Setup Complete.")

#loading the data

data_filepath_mtp = "anon_data_analysts_mtp.csv"
data_filepath_bdx = "anon_data_analysts_bdx.csv"
analysts_mtp = pd.read_csv(data_filepath_mtp)
analysts_bdx = pd.read_csv(data_filepath_bdx)

# Replace inf and -inf with NaN explicitly
analysts_mtp = analysts_mtp.replace([np.inf, -np.inf], np.nan)
```

```python
print("In Bordeaux ", analysts_bdx.shape[0],", people are found under Data Analysts on Linkedin.")

print("In Montpellier ", analysts_mtp.shape[0],", people are found under Data Analysts on Linkedin.")
```

```python
#Looking for empty cells in each column
missing_values_count_mtp = analysts_mtp.isnull().sum()
missing_values_count_mtp[:]
```

```python
missing_values_count_bdx = analysts_bdx.isnull().sum()
missing_values_count_bdx[:]
```

```python
#cleaning
#removing all the empty companies values
companies_mtp = analysts_mtp.dropna(subset=['company_name'])
companies_bdx = analysts_bdx.dropna(subset=['company_name'])
```

```python
check_missing_values_count = companies_mtp.isnull().sum()
check_missing_values_count[:]
#good
```

```python
graph_companies_mtp = companies_mtp['company_name'].value_counts()

# Filter to include only companies with more than 1 analyst
graph_companies_mtp = graph_companies_mtp[graph_companies_mtp > 1]

sns.histplot

# Plot the bar chart
plt.figure(figsize=(10, 6))
graph_companies_mtp.sort_values().plot(kind='barh')
# Add counts to the bars
for index, value in enumerate(graph_companies_mtp.sort_values()):
    plt.text(value, index, str(value), va='center')  # `value` is the count, `index` is the bar position
plt.title('Companies employing more than one data analyst in Montpellier according to Linkedin')
plt.xlabel('Count')
plt.ylabel('Name')
```

```python
graph_companies_bdx = companies_bdx['company_name'].value_counts()

# Filter to include only companies with more than 1 analyst
graph_companies_bdx = graph_companies_bdx[graph_companies_bdx > 1]

sns.histplot

# Plot the bar chart
plt.figure(figsize=(10, 14))
graph_companies_bdx.sort_values().plot(kind='barh')
# Add counts to the bars
for index, value in enumerate(graph_companies_bdx.sort_values()):
    plt.text(value, index, str(value), va='center')  # `value` is the count, `index` is the bar position
plt.title('Companies employing more than one data analyst in Bordeaux according to Linkedin')
plt.xlabel('Count')
plt.ylabel('Name')
```
